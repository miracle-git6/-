## Проверка гипотез

#### Органический трафик не отличается от платного с точки зрения CR (Conversion Rate) в целевые события.

```python
organic_mediums = ['organic', 'referral', '(none)']
df['is_organic'] = df['utm_medium'].isin(organic_mediums)

cr_organic = df[df['is_organic'] == True]['is_converted'].mean()
cr_paid = df[df['is_organic'] == False]['is_converted'].mean()
```

```python
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(
    df[df['is_organic']]['is_converted'],
    df[~df['is_organic']]['is_converted'],
    equal_var=False
)
```

```python
print(f"CR Organic: {cr_organic:.4f}")
print(f"CR Paid: {cr_paid:.4f}")
print(f"T-test: t={t_stat:.3f}, p={p_val:.4f}")
```

```python
# P-value < 0.05, значит отвергаем нулевую гипотезу. 
# Разница в CR между органическим и платным трафиком статистически значима.
# Органический трафик конвертируется лучше, чем платный.
```

#### Трафик с мобильных устройств не отличается от трафика с десктопных устройств с точки зрения CR (Conversion Rate) в целевые события.

```python
df['is_mobile'] = df['device_category'] == 'mobile'

cr_mobile = df[df['is_mobile']]['is_converted'].mean()
cr_desktop = df[~df['is_mobile']]['is_converted'].mean()
```

```python
t_stat, p_val = ttest_ind(
    df[df['is_mobile']]['is_converted'],
    df[~df['is_mobile']]['is_converted'],
    equal_var=False
)
```

```python
print(f"CR Mobile: {cr_mobile:.4f}")
print(f"CR Desktop: {cr_desktop:.4f}")
print(f"T-test: t={t_stat:.3f}, p={p_val:.4f}")
```

```python
# P-value < 0.05, значит нулевая гипотеза отвергается.
# Разница между CR мобильных и десктопных пользователей статистически значима.
# Пользователи с десктопов конвертируются лучше, чем с мобильных устройств.
```

#### Какие авто пользуются наибольшим спросом? У каких авто самый лучший показатель CR (Conversion Rate) в целевые события?

```python
pd.set_option('display.max_colwidth', None)
print(df_h['hit_page_path'].head(10))
```

```python
def extract_make_model(path):
    if pd.isna(path):
        return None
    try:
        start = path.find('/cars/')
        if start == -1:
            return None
        subpath = path[start+6:]
        if subpath.startswith('all/'):
            subpath = subpath[4:]
        subpath = subpath.split('?')[0]  
        parts = subpath.split('/')
        make = parts[0] if len(parts) > 0 else None
        model = parts[1] if len(parts) > 1 else None
        if make and model:
            return f"{make} {model}"
        elif make:
            return make
        else:
            return None
    except Exception:
        return None
```

```python
df_h['car_make_model'] = df_h['hit_page_path'].apply(extract_make_model)
print(df_h[['hit_page_path', 'car_make_model']].head(10))
```

```python
views = df_h['car_make_model'].value_counts().reset_index()
views.columns = ['car_make_model', 'views']
print(views.head(20))
```

```python
stats = df_h.groupby('car_make_model').agg(views=('session_id', 'count'),
    conversions=('is_conversion', 'sum')).reset_index()

stats['conversion_rate'] = stats['conversions'] / stats['views']
print(stats.sort_values('views', ascending=False).head(20))
```

```python
print(stats['car_make_model'].nunique())
```

```python
stats['length'] = stats['car_make_model'].apply(lambda x: len(str(x)))

print(stats['length'].describe())
print(stats[stats['length'] <= 8]['car_make_model'].unique()[:20])
print(stats[stats['length'] >= 20]['car_make_model'].unique()[:20])
```

##### Самый популярный спрос (по объёму трафика): skoda rapid(442779), lada vaz vesta(404150), volkswagen polo(318264). А самый лучший показатель CR: lada-vaz vesta, mercedes-benz e-klasse, volkswagen polo.
